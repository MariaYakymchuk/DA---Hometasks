4. Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
    
    а) дорівнюватиме семи;
    б) буде не меншим за 20;
    в) буде парним.

а) Кількість можливих результатів:
Оскільки кожен кубик має 6 граней (від 1 до 6), загальна кількість можливих результатів для двох кубиків дорівнює: 6*6 = 36

Шукаємо пари, де добуток чисел дорівнює 7. Таких пар не існує

P = 0/36 = 0

Отже, ймовірність того, що добуток чисел на двох кубиках дорівнюватиме 7, дорівнює 0.

б) Шукаємо пари, де добуток чисел ≥ 20. 

6×4=24
6×5=30
6×6=36
5×4=20
5×5=25
5×6=30
4×5=20
Отже, є 7 пар, де добуток чисел ≥20

P = 7/36 = 0, 19 = 19%

Отже, ймовірність того, що при кидку двох гральних кубиків добуток чисел на них буде не меншим за 20 є 19%

в) Всього кількість випадків 36. При множенні парного числа на парне та на непарне, результатом буде парне число. Тобто непарним числом в результаті множення буде тільки випадок, коли на обох кубиках випали непарні числа, 1, 3, 5. Таких випадків 9: 1*1, 1*3, 1*5, 3*1, 3*3, 3*5, 5*1, 5*3, 5*5. Отже випадків, коди добуток чисел, які випали на кубиках, є парним: 36 - 9 = 27.
P = 27/36 = 3/4 = 0,75

Отже, ймовірність того, що при кидку двох гральних кубиків добуток чисел буде парним є 75%
