1. Скільки різних паролів можна скласти, використовуючи три літери англійського алфавіту, якщо літери не повинні повторюватися?

Використовуємо "Перестановку".
Всього 3 позиції. 
n обєктів
n позицій

Pn = n * (n-1) * 1 = 3 * 2 * 1 = 6

Отже, у нас є 6 способів скласти пароль з трьох літер алфавіту так, щоб вони не повторювалися

2. Є група з 8 людей. Скількома способами можна обрати комітет із 3 осіб?

Використовуємо "Сполучення"

n = загальна кулькість людей = 8

r = кількість людей в комітеті = 3

Формула:

C(n,r) = n! / (n-r!)*r! 
C = 8! / (8-3)! * 3! = 8*7*6*5!/ 5! * (3*2*1) = 336 / 6 = 56

Отже, кількість способів обрати комітет із 3 осіб з 8 людей становить 56.

3.  Скількома різними способами можна розташувати на полиці 5 різних книг?

Всього 5 позиції. 
n обєктів
n позицій

Отже, у нас є 120 способів розташувати 5 різних вниг
